# Anatomy of a Class

## Definition

To create models of things that exist in the real world, blueprints are made into a class. A class then uses instances of a class, or objects, to represent indiviual examples of that blueprint in that system. 

Each object has its own set of data (attributes) and can perform actions (methods). This allows the object to interact with other objects. 

## Example


**Class:** Jobs



**Objects:**
1. Mom

    **Attributes:**
    * Title: Lawyer
    * Description: A lawyer is a professional who provides legal advice, represents clients in legal matters, and interprets the law.
    * Salary: $100,000
    
2. Dad

    **Attributes:**
    * Title: Software Engineer 
    * Description: A software engineer designs, develops, and maintains software systems and applications.
    * Salary: $130,000
